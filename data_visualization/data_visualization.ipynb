```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"F:\computational statistic\computational-statistic-for-sem3\dataset\train.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', alpha=0.6)
plt.title('Scatter Plot of Sale Price vs. Ground Living Area')
plt.xlabel('Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], bins=30, kde=True)
plt.title('Histogram of Sale Price')
plt.xlabel('Sale Price ($)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Neighborhood', y='SalePrice', estimator=sum, ci=None)
plt.title('Total Sale Price by Neighborhood')
plt.xticks(rotation=45)
plt.ylabel('Total Sale Price ($)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Neighborhood', y='SalePrice', hue='OverallQual', ci=None)
plt.title('Average Sale Price by Neighborhood and Overall Quality')
plt.xticks(rotation=45)
plt.ylabel('Average Sale Price ($)')
plt.grid()
plt.legend(title='Overall Quality')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Neighborhood', y='SalePrice')
plt.title('Box Plot of Sale Price by Neighborhood')
plt.xticks(rotation=45)
plt.ylabel('Sale Price ($)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Neighborhood', y='SalePrice', hue='OverallQual')
plt.title('Grouped Box Plot of Sale Price by Neighborhood and Overall Quality')
plt.xticks(rotation=45)
plt.ylabel('Sale Price ($)')
plt.grid()
plt.legend(title='Overall Quality')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Histogram
sns.histplot(df['SalePrice'], bins=30, kde=True, ax=ax, color='lightblue', alpha=0.5)

# Box plot
sns.boxplot(data=df, y='SalePrice', ax=ax, color='red', width=0.2)

ax.set_title('Box Plot Overlayed on Histogram of Sale Price')
ax.set_xlabel('Sale Price ($)')
ax.set_ylabel('Frequency / Box Plot')
ax.grid()

plt.show()
```

```python
# Selecting relevant numerical columns for pairwise plots
numerical_cols = ['SalePrice', 'GrLivArea', 'TotalBsmtSF', 'GarageCars']

sns.pairplot(df[numerical_cols])
plt.suptitle('Pairwise Plots of Selected Numerical Features', y=1.02)
plt.show()
```
