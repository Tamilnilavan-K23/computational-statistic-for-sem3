```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
df = (r"F:\computational statistic\computational-statistic-for-sem3\dataset\train.csv")
data = pd.read_csv(df, header = 0)
```

```python
threshold = 1.5
```

```python
observed_heads = (data['SalePrice'] > threshold).sum()
observed_tails = (data['SalePrice'] <= threshold).sum()
```

```python
prior_alpha, prior_beta = 2, 2
```

```python
posterior_alpha = prior_alpha + observed_heads
posterior_beta = prior_beta + observed_tails
```

```python
x = np.linspace(0, 1, 100)
prior_dist = beta.pdf(x, prior_alpha, prior_beta)
posterior_dist = beta.pdf(x, posterior_alpha, posterior_beta)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, prior_dist, label=f'Prior (Beta({prior_alpha},{prior_beta}))', color='blue', linestyle='--')
plt.plot(x, posterior_dist, label=f'Posterior (Beta({posterior_alpha},{posterior_beta}))', color='green', linestyle='-')
plt.xlabel('Probability of Sales in the Dataset')
plt.ylabel('Density')
plt.legend(loc='best')
plt.title("Bayesian Inference for Sales Probability using Conjugate Prior")
plt.grid(True)
plt.show()
```

```python
print(f"Observed Heads (Discount > Treshold): {observed_heads}")
print(f"Observed Tails (Discount <= Treshold): {observed_tails}")
print(f"Posterior Alpha: {posterior_alpha}, Posterior Beta: {posterior_beta}")
```
